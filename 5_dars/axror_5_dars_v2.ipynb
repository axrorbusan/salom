```python
import pandas as pd
df = pd.read_csv('personal_transactions.csv')

from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
#bolmasam birinchi label encoding amalga oshiramiz category bilan
df['Category']=encoder.fit_transform(df['Category'])
df.info()
```

```python
df.head(10)
```

```python
#endi one hot encoding bilan account name ishlatamiz
dudunamber=pd.get_dummies(df['Account Name'], prefix='acc', dtype=int)
dudunamber
```

```python
df.info()
```

```python
#endi Account Name column ni ochirib orniga dudunamber imni qoshib merge qilaman
df=pd.concat([df.drop(columns='Account Name'), dudunamber], axis=1)
df.info()
```
