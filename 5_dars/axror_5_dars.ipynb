```python
import pandas as pd
df = pd.read_csv('Finance_data.csv')
df.info()
```

```python
damdam=pd.get_dummies(df['gender'], prefix='jins', dtype=int)
damdam
```

```python
df.info()
```

```python
df=pd.concat([df.drop(columns='gender'), damdam], axis=1)
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
encoder
```

```python
df.info()
```

```python
df['Investment_Avenues']=encoder.fit_transform(df['Investment_Avenues'])
df.info()
```
